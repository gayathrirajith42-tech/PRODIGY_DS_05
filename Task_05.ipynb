```python
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
print("openpyxl installed successfully")
df = pd.read_excel("Road Accident Data.xlsx", engine="openpyxl")
```

```python
print(df.columns)
```

```python
# Display first 5 rows
print(df.head())
```

```python
# Missing values
print(df.isnull().sum())
```

```python
df['Accident_Severity'].value_counts()
```

```python
df['Accident_Severity'].value_counts().plot(kind='bar')
plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
df['Weather_Conditions'].value_counts().head(10).plot(kind='bar')
plt.title('Weather Conditions During Accidents')
plt.show()
```

```python
df['Road_Surface_Conditions'].value_counts().plot(kind='bar')
plt.title('Road Surface Conditions')
plt.show()
```

```python
df["Hour"] = df["Time"].apply(lambda x: x.hour if pd.notnull(x) else None)
```

```python
df[["Time","Hour"]].head(10)
```

```python
df['Hour'].value_counts().sort_index().plot(kind='line')
plt.title('Accidents by Hour')
plt.show()
```

```python
df['Accident_Severity'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Severity Percentage')
plt.ylabel('')
plt.show()
```
